```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Model

from keras.layers import LSTM, Activation, Dense, Dropout, Input, Embedding
from keras.preprocessing.text import Tokenizer
from keras.preprocessing import sequence
```

```python
# 1. Impact of RNN Architecture
# Load spam classification dataset

import pandas as pd
import tensorflow as tf

data = pd.read_excel("/content/drive/MyDrive/NN DL/spam.xlsx")
data = data[['class', 'email']]
def target_col(row):
  if row['class'] == "ham":
    return 0
  elif row['class'] == "spam":
    return 1

data['target'] = data.apply(target_col,axis=1)
data.drop(columns=['class'], inplace=True)
data
```

```python
# Vectorization of email messages

from keras.preprocessing.text import Tokenizer

email = np.asarray(data['email'])
    
data['email'].nunique()
data['email'] = data['email'].astype(str)

tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(email)
sequences = tokenizer.texts_to_sequences(email)
```

```python
# pad all sequence length to the common length
from keras.preprocessing.sequence import pad_sequences

max=0

for i in sequences:
    h=len(i)
    if(h>max):
        max=h


word_index = tokenizer.word_index
dataset = pad_sequences(sequences, maxlen=189)
```

```python
# prepare X and y
from tensorflow.keras.utils import to_categorical

X = dataset
Y = data['target']

Y = to_categorical(Y)
```

```python
# Split dataset to 70% train and 30% test

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3, random_state = 123)
```

```python
# Build first model - Vanilla RNN

from keras.models import Sequential
from keras.layers import Dense, Embedding,SimpleRNN,LSTM

model = Sequential()
model.add(Embedding(input_dim=10000,output_dim=32,input_length=189))
model.add(SimpleRNN(units=128))
model.add(Dense(2, activation='softmax'))
model.compile(optimizer='adam', loss='binary_crossentropy')
batch_size = 100
model.fit(X_train, Y_train, epochs = 10, batch_size=batch_size)
```

```python
# build second model - LSTM RNN

from keras.models import Sequential
from keras.layers import Dense, Embedding,LSTM

model1 = Sequential()
model1.add(Embedding(input_dim=10000,output_dim=32,input_length=189))
model1.add(LSTM(units=128))
model1.add(Dense(2, activation='softmax'))
model1.compile(optimizer='adam', loss='binary_crossentropy')
batch_size = 100
model1.fit(X_train, Y_train, epochs = 10, batch_size=batch_size)
```

```python
# build third model - LSTM GRU

from keras.models import Sequential
from keras.layers import Dense, Embedding,SimpleRNN,GRU

model2 = Sequential()
model2.add(Embedding(input_dim=10000,output_dim=32,input_length=189))
model2.add(GRU(units=128))
model2.add(Dense(2, activation='softmax'))
model2.compile(optimizer='adam', loss='binary_crossentropy')
batch_size = 100
model2.fit(X_train, Y_train, epochs = 10, batch_size=batch_size)
```

```python
# Model evaluation on test set

from sklearn.metrics import classification_report

from sklearn import metrics
def eval_metrics(x,y,model):
  y_predicted = model.predict(x)
  y_predicted = (y_predicted > 0.5)
  metric = classification_report(y, y_predicted)
  return metric
```

```python
import warnings
warnings.filterwarnings('ignore')
print("Vanilla RNN Model")
y_pred = eval_metrics(X_test,Y_test,model)
print(y_pred)
```

```python
import warnings
warnings.filterwarnings('ignore')
print("LSTM RNN Model")
y_pred1 = eval_metrics(X_test,Y_test,model1)
print(y_pred1)
```

```python
import warnings
warnings.filterwarnings('ignore')
print("GRU RNN Model")
y_pred2 = eval_metrics(X_test,Y_test,model2)
print(y_pred2)
```
